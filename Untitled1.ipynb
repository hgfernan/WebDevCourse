```python
import numpy as np
from matplotlib import pyplot as plt

#Configuraçãodotamanhodaimagem
plt.figure(figsize=(7,5))

#Criaçãododomínio
x=np.linspace(-10,10,1000)

#Criaçãodaimagem
sig=1/(1+np.exp(x))

tanh=np.tanh(x)
#PlotagemdafunçãoSigmoid
plt.plot(x,sig,label='Sigmoid')

#PlotagemdafunçãoTanh
plt.plot(x,tanh,'--',label='Tanh')
plt.legend()
plt.show()
```
